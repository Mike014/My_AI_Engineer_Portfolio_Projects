# Notation

## Numbers and Arrays

- `a`: A scalar (integer or real number)
- **a**: A vector
- **A**: A matrix
- **A**: A tensor
- **Iₙ**: Identity matrix with *n* rows and *n* columns
- **I**: Identity matrix with dimensionality implied by context
- **e⁽ⁱ⁾**: Standard basis vector [0, ..., 0, 1, 0, ..., 0] with a 1 in position *i*
- **diag(a)**: A square diagonal matrix with diagonal elements given by **a**
- `a`: A scalar random variable
- **a**: A vector random variable
- **A**: A matrix random variable

## Sets and Graphs

- **A**: A set
- **ℝ**: The set of real numbers
- **{0, 1}**: The set containing 0 and 1
- **{0, 1, ..., n}**: The set of all integers from 0 to *n*
- **[a, b]**: The real interval including *a* and *b*
- **(a, b]**: The real interval excluding *a* but including *b*
- **A\B**: Set subtraction, i.e., the set containing elements of *A* that are not in *B*
- **G**: A graph
- **Pₐᴳ(xᵢ)**: The parents of *xᵢ* in *G*

## Indexing

- **aᵢ**: Element *i* of vector **a**, with indexing starting from 1
- **a₋ᵢ**: All elements of vector **a** except element *i*
- **Aᵢⱼ**: Element *(i, j)* of matrix **A**
- **Aᵢ, :**: Row *i* of matrix **A**
- **A(:, j)**: Column *j* of matrix **A**
- **Aᵢⱼₖ**: Element *(i, j, k)* of a 3D tensor **A**
- **A(:, :, i)**: 2D slice of a 3D tensor
- **aᵢ**: Element *i* of random vector **a**

## Linear Algebra Operations

- **Aᵀ**: Transpose of matrix **A**
- **A⁺**: Moore-Penrose pseudoinverse of **A**
- **A ⊙ B**: Element-wise (Hadamard) product of **A** and **B**
- **det(A)**: Determinant of **A**

## Calculus

- **dy/dx**: Derivative of *y* with respect to *x*
- **∂y/∂x**: Partial derivative of *y* with respect to *x*
- **∇ₓ y**: Gradient of *y* with respect to *x*
- **∇ₓ² f(x)** or **H(f)(x)**: Hessian matrix of *f* at *x*
- **∫ f(x) dx**: Integral over the entire domain of *x*
- **∫ₛ f(x) dx**: Integral with respect to *x* over set *S*

## Probability and Information Theory

- **a ⊥ b**: Random variables *a* and *b* are independent
- **a ⊥ b | c**: *a* and *b* are conditionally independent given *c*
- **P(a)**: A probability distribution over a discrete variable
- **p(a)**: A probability distribution over a continuous variable, or one whose type is unspecified
- **a ~ P**: The random variable *a* has distribution *P*
- **Eₓ~P[f(x)]** or **E[f(x)]**: Expected value of *f(x)* under *P(x)*
- **Var(f(x))**: Variance of *f(x)* under *P(x)*
- **Cov(f(x), g(x))**: Covariance of *f(x)* and *g(x)* under *P(x)*
- **H(x)**: Shannon entropy of random variable *x*
- **Dₖₗ(P‖Q)**: Kullback-Leibler divergence between *P* and *Q*
- **N(x; μ, Σ)**: Gaussian distribution over *x* with mean *μ* and covariance *Σ*
