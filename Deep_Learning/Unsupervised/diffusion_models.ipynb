# **Diffusion models**

A **class of generative machine learning models** that produce **high-quality synthetic data**.
These models work by gradually **transforming a noisy initial sample** through a **series of passes**.

**Diffusion models work by defining two processes**: a **forward process that adds noise** to the data, and a **reverse process that learns to remove the noise** step by step to **reconstruct the original data**. This mechanism allows for the generation of **high-quality samples from random noise**.

- **Synthetic data**: This is **data that is artificially generated by the model**, not from real measurements or observations.
- **Noise**: In the scattering process, noise is a **random disturbance added to the data**
